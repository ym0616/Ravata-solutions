# voltage split function

```python
def split_dataframe_voltage(filename): 
    # read file
    df = pd.read_csv(filename)
    # groupby by voltage
    grouped = df.groupby('Voltage')
    data = pd.DataFrame()
    # get the index of grouped datafreame
    a = grouped.count()
    # loop through all the indices, which are the different voltages
    for i in a.index:
        # create data dictionary, keys = voltage; values = dataframe that corresponds to each assigned voltage
        data = {vol:df[df['Voltage']==vol] for vol in set(df['Voltage'])}
    return data
```

```python
#call the function
result = split_dataframe_voltage('Well__2.txt')
# result['10 MV'],result['20 MV'], and so on
result["10 MV"]
```

# Frequency split function

```python
def split_dataframe_frequency(result):
    for j in result.keys():
        # create voltage dictionary, keys = frequency; values = dataframe that corresponds to each assigned voltage
        freq = {fre:result[result['Frequency']==fre] for fre in set(result['Frequency'])}
        
    return freq
```

```python
# must put result['10 MV'], this format! 
sub_result = split_dataframe_frequency(result['10 MV'])
# short for result['10 MV']['130KHZ']
sub_result['130KHZ']
```
